```python
DOMAIN = 'https://gatewayapi.prodam.sp.gov.br:443/financas/orcamento/sof/v3.0.1/'
```

```python
AUTH = 'e085f433268dba68d07e55fcd9cd6fe'
```

```python
def get_json(url, auth=AUTH):
    
    with requests.get(url, headers={'Authorization' : f'Bearer {auth}'}) as r:
        assert r.status_code == 200
        
        return r.json()
```

```python
def get_orgaos(ano):
    
    params = f"orgaos?anoExercicio={ano}"
    
    url = DOMAIN + params
    
    return get_json(url)
```

```python
def parse_orgaos(ano):
    
    resp_orgaos = get_orgaos(ano)
    lst_orgaos = resp_orgaos['lstOrgaos']
    df_orgaos = pd.DataFrame(lst_orgaos)
    df_orgaos['ano'] = ano
    
    return df_orgaos
```

```python
orgaos = parse_orgaos(2022)
```

```python
orgaos
```

```python
def solve_mes(ano):
    '''Verifica se o ano é o atual.
    Se for, retorna mês atual. Se não, retorna mês 12.'''
    
    hoje = datetime.today()
    ano_atual = hoje.year
    if ano == ano_atual:
        return hoje.month
    elif ano < ano_atual:
        return 12
    else:
        return ValueError(f'Ano {ano} no futuro!')
```

```python
def get_despesas(cd_orgao: str, ano: int):
    
    mes = solve_mes(ano)
    
    params = (f"despesas?anoDotacao={ano}&"
              f"mesDotacao={mes}&"
              f"codOrgao={cd_orgao}")
    
    url = DOMAIN + params
    
    return get_json(url)
```

```python
def parse_despesas(cd_orgao: str, ano: int):
    
    resp_despesas = get_despesas(cd_orgao, ano)
    #apesar do nome, é um dicionário
    dici_desp = resp_despesas['lstDespesas']
    
    keys = [
        'valOrcadoInicial',
        'valOrcadoAtualizado',
        'valDisponivel',
        'valEmpenhadoLiquido',
        'valLiquidado'
    ]
    
    if dici_desp is None:
        return {key : None for key in keys}
    
    return {key : dici_desp[key] for key in keys}
```

```python
def all_despesas(lst_orgaos: list, ano: int):
    
    despesas = []
    for cd_orgao in lst_orgaos:
        desp = parse_despesas(cd_orgao, ano)
        desp['cd_orgao'] = cd_orgao
        despesas.append(desp)
        
    return pd.DataFrame(despesas)
```

```python
df_despesas = all_despesas(orgaos['codOrgao'], ano=2022)
```

```python
merged = pd.merge(df_despesas, orgaos, left_on='cd_orgao', right_on='codOrgao',
        how='inner')
```
